```python
topWeights = []
middleWeights = []
bottomWeights = []
for i in range(len(reshapedWeights)):
    topWeights = np.append(topWeights, reshapedWeights[i][0])
    middleWeights = np.append(middleWeights, reshapedWeights[i][1])
    bottomWeights = np.append(bottomWeights, reshapedWeights[i][2])
```

```python
print(topWeights.shape[0])
```

```python
finalWeights = np.append(topWeights, np.append(middleWeights, bottomWeights)).reshape(3, 12288)
print(finalWeights)
```

```python
#final shape needs to be 192x192
testMatrix = np.zeros((192, 192))
for i in range(64):
    testMatrix[i*3] = finalWeights[0][i*192:(i+1)*192]
    testMatrix[i*3+1] = finalWeights[1][i*192:(i+1)*192]
    testMatrix[i*3+2] = finalWeights[2][i*192:(i+1)*192]
```

```python
plt.figure(figsize=(15,15))
ax = sns.heatmap(testMatrix, xticklabels=False, yticklabels=False, cmap="Blues", square=True)
ax.savefig("output.png")
```
